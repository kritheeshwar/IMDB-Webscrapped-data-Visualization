```python
#set up chrome options 
options = Options()                                        #initializes an instance of options, allows to set various settings or configurations for the chrome browsers
#option.add_arguments("--headless")                        #run chrome in  headless mode (without opening a window)
options.add_argument("--disable-gpu")                      #Disable GPU acceleration (optional)

#initialize webdriver
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options) #The driver object represents the browser and can be used to interact with web pages.

#open the flipkart mobile phones store page
url = "https://www.imdb.com/chart/toptv/?ref_=nv_tvv_250"
driver.get(url) #instructs the browser to navigate to the specified URL

#wait for the page to load
time.sleep(5)

content = driver.page_source #retrives the entire HTml source of the loaded web page and stores it in the content variable.
soup = BeautifulSoup(content)
tvshows = soup.find_all('li', class_="ipc-metadata-list-summary-item sc-4929eaf6-0 DLYcv cli-parent")
```

```python

import pandas as pd
import re

# Load the Excel file
file_path = 'output7.xlsx'
df = pd.read_excel(file_path)

# Define a function to split Years, Episodes, and Rating
def split_info(row):
    parts = row.split(" ")
    # Extract Years (first part, which may contain a hyphen)
    years = parts[0]
    
    # Extract Episodes (next part, before 'eps')
    episodes = parts[1]
    
    # Join the rest as Rating
    rating = " ".join(parts[3:])  # Skip 'eps' and take the rest as rating
    
    return pd.Series([years, episodes, rating])

# Apply the function to each row in the DataFrame and create new columns 'Years', 'Episodes', 'Rating'
df[['Years', 'Episodes', 'Rating']] = df['Years_Category_Episodes'].apply(split_info)

# Drop the original 'Years_Category_Episodes' column if needed
df = df.drop(columns=["Years_Category_Episodes"])

# Display the resulting DataFrame
print(df)

# Save the modified DataFrame back to an Excel file
#output_file_path = '/mnt/data/split_output_alternative.xlsx'
#df.to_excel(output_file_path, index=False)

```

```python
# Define a function to split the number and title
def split_number_title(row):
    match = re.match(r"(\d+)\.\s*(.+)", row)
    if match:
        return pd.Series([match.group(1), match.group(2)])
    return pd.Series([None, row])

# Apply the function to the DataFrame
df[['Number', 'Title']] = df['show_names'].apply(split_number_title)

# Drop the original 'show_names' column if you don't need it
df = df.drop(columns=['show_names'])

# Display the resulting DataFrame
print(df)
```

# ONE SAMPLE T TEST

```python
from scipy.stats import ttest_1samp
hm = 9.5  # HYPOTHESISED MEAN 
t_stat, p_value = ttest_1samp(df['Ratings'],hm)
alpha = 0.05 # acceptance error percent
print(f"T-statistic:{t_stat}")
print(f"P-value:{p_value}")
if p_value<alpha:
    print(f"Reject null hypothesis at alpha ={alpha}")
else:
    print(f"Faild to reject null hypothesis at alpha = {alpha}")
    
```

# output suggests that the t-statistic is approximately -51.87 and the p-value is extremely small, 1.75e-135. Given the p-value is much less than 0.05,
# so we reject the null hypothesis at the 5% significance level.

# This indicates that the mean of the Ratings  significantly differs from the hypothesized mean (9.5).
